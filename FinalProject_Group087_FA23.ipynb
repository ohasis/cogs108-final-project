# Abstract

The topic of gaming has gained immense popularity in recent years, and video game genres serve as foundational frameworks that can have significant impacts on player experiences and expectations. Exploring whether certain genres generate more demand has always been a subject of interest for game developers, publishers, and stakeholders. Discovering any patterns between specific genres and success would likely have a strong effect on the direction gaming takes in the foreseeable future, as companies would use this information to make decisions regarding budgeting, marketing, and game design. Thus, our group decided it would be useful to analyze how significant of an impact a video game’s genre or genres has on its success.

To conduct this analysis, we collected data on thousands of different video games from the last few decades, along with reviews, ratings, global copies sold, and critic/user scores. These provided good indicators as to whether a game was well-liked and could be considered “successful.” We operationalized success as the positivity of a game’s reviews, the ratings of its critics/users, and its overall number of sales. These factors were determined to be the dependent variables, while the video game genres were seen as the independent variables. With this relationship in mind, we could map the data to draw comparisons between genres and success. Through synthesizing and analyzing the chosen data sets, the following was discovered:

Initially, the results of our ANOVA tests led us to believe that genre did have an impact on a video game's success. However, while some genres showed positive correlations between sentiment scores and acclaim, this didn't always translate to increased sales. This complexity highlights that genre isn't the sole driver of a game's success, emphasizing the need to consider broader factors like marketing tactics and audience trends. While genres influence ratings, their direct impact on sales is less straightforward. This goes against the assumption that games that have been well-received in popular genres before will automatically achieve strong sales. This emphasizes the importance of developers’ consideration of a multitude of factors beyond genre when aiming for commercial success in the gaming industry.

# Research Question

**Are there certain combinations of video game genres that consistently earn more positive reviews, higher critic and user scores and an increased number of copies sold? Can genres be feasibly used as predictors for success as measured by these three variables?**

## Background and Prior Work

The gaming industry has seen tremendous growth over the past few decades, with a diverse selection of genres available to players all over the world. The reception of these games, both initial and long-term, which are reflected in user and critic reviews, along with their sales performance is of significant interest to the developers creating these games, as well as the gaming community which depends on many of these statistics to make their decisions on what to purchase or play. Understanding how these factors interact, particularly across different game genres, is in the best interest of developers and publishers aiming to align their products with market demands and player preferences. 

Regarding previous research work that has been performed in this domain, previous work has primarily been focused on analyzing the impact of reviews on sales. One particular study, a research study performed by Feray Adigüzel of Nottingham Trent University in September of 2021 suggests a strong correlation between positive reviews of games through formats such as YouTube videos, and to what extent these opinions shared with large audiences affect the sales of different games <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). Additionally, the project “Data Trends in Video Game Sales and Ratings” created by Casey Hoffman in 2021 <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) conducted an analysis of multiple research questions in the same space, most relevantly “Are user and critic ratings correlated, for different genres?” and “Do either user or critic ratings correlate with game sales?” Both of these questions address related questions to ours; however, they do not discuss our question: how do genre combinations affect the relationship between review scores and sales numbers?

For some more context, research published by researchers from Illinois Wesleyan University <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) explored various determinants of video game software sales. While it didn't specifically discuss genre combinations, it laid a foundation for understanding consumer preferences and how they translate to sales, which could be instrumental in analyzing how genre combinations might sway consumer perceptions and sales. Similarly, a report by a researcher at Duke University in 2019 <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) aimed to identify factors influencing the success of video game sales. It provides a broad framework within which the impact of genre combinations could be analyzed, although it doesn’t specifically delve into genre combinations. This work could serve as a reference point for structuring a more focused investigation into how genre combinations affect sales performance​

In terms of further analysis in this field, the insights derived from potential further research might be instrumental in guiding developers towards creating games that not only resonate with their target audience but also understand which characteristics tend to lead to commercial success while doing so. The discussed data shows how game genres, reviews, and sales are connected, but not much is found about how mixing genres affects these areas. Our analysis aims to fill this gap. We want to explore how combining genres impacts both user and critic reviews, and how this affects sales. Our analysis seeks to add a new angle to the existing information, helping to better understand how genre combinations work in the gaming market.

# Hypothesis


We predict that genre popularity (operationalized as the aggregate total of sales of titles within that particular genre, the most number of positive reviews in comparison to the other genres in the sample, and the highest overall game ratings) will generate an effect on the relationship between these variables and their independent variables: critic and user scores and copies of games sold within that particular genre. 

This is based on our theory that video game titles with mass-market appeal will belong to certain genres wherein the critical reviews are more correlated to hypothesized commercial appeal rather than more niche genres where critics will only be interested in reviewing if they have a pre-existing interest in the genre, potentially diverging their reviews from sales.

### Reviews DF Setup

```python
df_reviews = pd.read_csv('data/popular_games_user_reviews.csv') 
df_reviews.columns
```

```python
df_reviews = df_reviews.drop_duplicates(['Title', 'Release Date'])
df_reviews = df_reviews.dropna(subset=['Title', 'Reviews', 'Rating'])

df_reviews = df_reviews[['Title', 'Release Date', 'Team', 'Rating', 'Times Listed', 'Number of Reviews', 
                         'Genres', 'Summary', 'Reviews', 'Plays', 'Playing', 'Backlogs', 'Wishlist']]

df_reviews['Release Date'] = pd.to_datetime(df_reviews['Release Date'], errors='coerce')
df_reviews['Rating'] = pd.to_numeric(df_reviews['Rating'], errors='coerce')

# Drop rows with NaT in 'Release Date'
df_reviews = df_reviews.dropna(subset=['Release Date', 'Rating'])

# Convert 'Release Date' to just the year
df_reviews['Release Date'] = df_reviews['Release Date'].dt.year

df_reviews.head(3)
```

```python
# Standardizing the columns of quantitative data from strings to floats

def num_convert(string):
    string = str(string)
    string = string.lower()
    string = string.strip()
    if 'k' in string:
        string = string.replace('k', '')
        return float(string) * 1000
    else:
        return float(string)

columns_to_convert = ['Times Listed', 'Number of Reviews', 'Plays', 'Playing', 'Backlogs', 'Wishlist']

for column in columns_to_convert:
    df_reviews[column] = df_reviews[column].apply(num_convert)
```

### Merging the Data

Now that we have two different dataframes, `df_reviews` and `df_sales` which we have loaded from their respective csv files and done some initial cleaning, we will want to merge both of these dataframes into one larger dataframe to make it easier to perform our analyses. In order to do this, we specify the columns below from each dataframe which we want to keep in our `merged_df`. For this analysis, we will only keep columns relevent to answering our research question. This includes `Title, Release Date, User Rating, Genres, Reviews, Global Sales, and Critic Score`. 

```python
reviews_columns = ['Title', 'Rating', 'Genres', 'Reviews', 'Release Date']
df_reviews_data = df_reviews[reviews_columns]
df_reviews_data.shape
```

```python
sales_columns = ['Title', 'Release Date', 'Global_Sales', 'Critic_Score']
df_sales_data = df_sales[sales_columns]
df_sales_data.shape
```

Once we have picked the columns we want to keep from each dataframe, we can merge both as shown below. We want a way to match up corresponding games in each dataframe, and we know the easiest way to accomplish this task will be by the title. The only issue is, different versions of different games can have the same title. In order to avoid this issue, we can also merge using the Release Date, which will make sure that we are mathcing up the right versions of the right games during our merge.

```python
merged_df = pd.merge(df_reviews_data, df_sales_data, on=['Title', 'Release Date'], how='inner')
merged_df = merged_df.dropna().reset_index(drop=True)
merged_df
```

```python
# Remove duplicate rows based on 'Title' and 'Release Date'
merged_df = merged_df.drop_duplicates(subset=['Title', 'Release Date'])
# Rename the 'Global_Sales' column to 'Global Sales'
merged_df = merged_df.rename(columns={'Global_Sales': 'Global Sales', 'Critic_Score': 'Metacritic Critic Score', 'Rating': 'Backloggd User Rating'})

# Define a function to convert a string representation of a list into a list
def convert_string_to_list(s):
    try:
        return ast.literal_eval(s)
    except (ValueError, SyntaxError):
        return []

# Apply the function to the 'Reviews' column
merged_df['Reviews'] = merged_df['Reviews'].apply(convert_string_to_list)

# Reset the index
merged_df = merged_df.reset_index(drop=True)
merged_df
```

To begin to look into the relationships that different genres have on the predictive variables that we have outlined, we want a better way to quantify which games belong to which genres, as a game can belong to multiple genres. Currently, we only have the column `Genres` which contains a list of strings of all the genres that a certain game belongs to, but this will be difficult to parse when performing our deeper analysis. To fix this issue, we can implement something similar to a one-hot-encoding system, where each possible genre will have its own column in the dataframe, and the value will either be a `1`, indicating the game is considered a part of that genre, or a `0` indicating it is not. This also resembles the quantification of a boolean system, where `True` might be quantified as `1`, while a `False` might be quantified as `0`.

```python
# Convert 'Genres' to string values and strip unwanted characters
merged_df['Genres'] = merged_df['Genres'].astype(str).str.strip("[]").str.replace("'", "")

# Create dummy variables for each genre
genres_dummies = merged_df['Genres'].str.get_dummies(sep=', ')
merged_df = pd.concat([merged_df, genres_dummies], axis=1)

merged_df.head()
```

## Statistical Analysis Via ANOVA Test

In the pursuit of understanding the dynamics of video game success across different genres, we employ the ANOVA (Analysis of Variance) statistical test. The decision to employ ANOVA comes from our objective to determine whether there are statistically significant differences in the performance metrics—such as user ratings, critic scores, and global sales among various game genres. Our hypothesis predicts that the genre of a game is a defining factor in a game's success, and the ANOVA test is a good method to test this hypothesis across multiple genres simultaneously.

By comparing our variables of interest across different genres, ANOVA helps us identify if any particular genre stands out in terms of higher ratings or sales. This is particularly insightful as it accounts for the variability within group means, which could be hard to notice when looking at each genre in isolation.

The result of this test will help us understand the impact of genre on a game's market performance. If the ANOVA test indicates significant differences, it would suggest that the game genre could be a predictor of success. This would not only validate the influence of genre on performance metrics but also give us further hints as to what genres to continue performing deeper analysis on.

In order to actually conduct our ANOVA tests, for each genre, we first split up the data that we have into two groups: games that belong to the genre and games that do not. Next, for each of the three variables we are considering from our research question, we perform an ANOVA test between both groups. The results of the ANOVA test, including the genre, variable, F-value, and p-value, are then stored in a Dataframe.

```python
# Get a list of genres
genres = genres_dummies.columns

# Create an empty DataFrame to store the results
anova_results = pd.DataFrame(columns=['Genre', 'Variable', 'F-value', 'p-value'])

# Perform ANOVA for each genre
for genre in genres:
    # Create two groups: games of this genre and games not of this genre
    group1 = merged_df[merged_df[genre] == 1]
    group2 = merged_df[merged_df[genre] == 0]
    
    # Perform ANOVA for each variable
    for variable in ['Backloggd User Rating', 'Global Sales', 'Metacritic Critic Score']:
        f_val, p_val = f_oneway(group1[variable], group2[variable])
        
        # Create a DataFrame for the results and concatenate it with the existing results
        result = pd.DataFrame({'Genre': [genre], 'Variable': [variable], 'F-value': [f_val], 'p-value': [p_val]})
        anova_results = pd.concat([anova_results, result], ignore_index=True)

anova_results
```

### ANOVA Test Visualization

Now that we have completed our tests, we have the relevant p-values for each genre and each of the three predictive variables. We want to consider a `95%` confidence interval for our test, and therefore can plot the p-values and a threshold at `p = 0.05` to be able to see which genres and predictive variable combos are statistically significant.

```python
variables = ['Backloggd User Rating', 'Global Sales', 'Metacritic Critic Score']

palette = sns.color_palette("tab20", len(anova_results['Genre'].unique()))

for variable in variables:
    # Filter the results for the current variable
    filtered_results = anova_results[anova_results['Variable'] == variable]
    
    # Sort the results by the p-value
    filtered_results = filtered_results.sort_values('p-value')
    
    color_map = dict(zip(filtered_results['Genre'].unique(), palette))
    
    # Create a bar plot for the p-values
    plt.figure(figsize=(10, 5))
    plt.bar(filtered_results['Genre'], filtered_results['p-value'], color=filtered_results['Genre'].map(color_map))
    plt.axhline(y=0.05, color='r', linestyle='--')  # Add a line at p=0.05
    plt.title(f'p-values for {variable} by Genre')
    plt.xlabel('Genre')
    plt.ylabel('p-value')
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Convert the significant results into a list of tuples
significant_pairs = anova_results[anova_results['p-value'] < 0.05][['Genre', 'Variable']].values.tolist()
significant_pairs = [tuple(x) for x in significant_pairs]

print(significant_pairs)
```

The provided graphs illustrate the p-values obtained from the ANOVA tests conducted to evaluate the effect of video game genres on Metacritic Critic Scores, Global Sales, and Backloggd User Ratings. The red dashed line in each graph represents the threshold p-value of 0.05, below which the differences in group means are considered statistically significant.

For the Metacritic Critic Score, it appears that certain genres have p-values below the `0.05` threshold, indicating that there are significant differences in critic scores among these genres. This suggests that the genre of a game could be a predictor of its critical reception, with some genres tending to receive higher or lower critic scores than others.

## Sentiment Analysis on Reviews to Find Statistically Significant Genres

Now that we have an idea of which genres seem to be statisticaly significant and relevant in affecting a certain predictive variable (Global Sales, Critic Scores, and User Ratings), we want to see if we can utilize the actual reviews given by users of each game to get a better understanding of how these reviews for different genres affect the variables of interest.

In order to do this, we performed some basic sentiment analysis using NLTK for each game, and assigned each game a sentiment score, between `-1` for the maximum negative review, to `1` for a maximum positive review. Using these sentiment scores, we can look at the statistically significant (genre, variable) pairs from the ANOVA test and see if the sentiment analysis can provide us a better understanding of their relationship.

```python
# Example to diagnose the format of reviews
sample_reviews = merged_df['Reviews'].iloc[0]  # Get reviews for the first game

print(sample_reviews)

# Check the first review
print("First review:", sample_reviews[0])
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download the VADER lexicon
nltk.download('vader_lexicon')

def preprocess_review(review):
    # Basic preprocessing: lowercasing and removing non-alphanumeric characters
    return ''.join([char.lower() if char.isalnum() or char.isspace() else ' ' for char in review])

def analyze_sentiment(reviews):
    sia = SentimentIntensityAnalyzer()
    compound_scores = []

    for review in reviews:
        preprocessed = preprocess_review(review)
        score = sia.polarity_scores(preprocessed)['compound']
        compound_scores.append(score)

    # Calculate the average compound score for the game
    return sum(compound_scores) / len(compound_scores) if compound_scores else 0

merged_df['sentiment_score'] = merged_df['Reviews'].apply(analyze_sentiment)
```

# Ethics & Privacy

- The data files that our group used are publicly available on Kaggle, which is a publicly accessible site.
- For the games in question, players must sign end-user license agreements before playing.
    - This measure establishes an agreement between the player and developer to ensure that the relative popularity of the game—for example, the number of downloads—is assessable and reviewers may translate it into a numerical score.
- To expound on this fact, the potential bias that may have occurred when the data was collected includes the following:
    - Customers who did not have a good experience with the video game are more likely to leave a review than those who had a good experience.
    - More casual players may leave rational reviews comprehensive by the general public, while more frequent players of other genres may leave irrational reviews.
    - Encountering other players’ reviews may have affected the players' review whether it is positive or negative.
- Data variation may be unavoidable when analyzing data since that would lead to the collection of user-specific data which introduces privacy concerns.

**How can we counteract these issues?**

Without knowing more about the game reviewers themselves, it is difficult to limit the biases that may have affected their reviews. However, by taking a random sample of the reviews, we are working to minimize the effect of these biases in our dataset. For reviews in other languages, our NLP methods could be used to translate these reviews to English before performing sentiment analysis to avoid completely ignoring reviews from different languages, as these reviews may offer insights into opinions from different countries and markets.
